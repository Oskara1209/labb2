```python
import numpy as np

# ersätt ".." med NaN och konvertera till float
df_svenska_som_andraspråk["Totalt (Poäng)"] = df_svenska_som_andraspråk["Totalt (Poäng)"].replace("..", np.nan)
df_svenska_som_andraspråk["Totalt (Poäng)"] = pd.to_numeric(df_svenska_som_andraspråk["Totalt (Poäng)"], errors='coerce')

# kontrollera igen
print(df_svenska_som_andraspråk[["Huvudman", "Totalt (Poäng)"]])
```

```python
import os
import matplotlib.pyplot as plt

# skapa en undermapp 'visualiseringar' om den inte finns 
# (Detta tog jag hjälp av chat gpt för att förstå, os är en inbyggd python modell som hjälper att skapa och lagra mappar och filer med mera)
output_dir = "visualiseringar"
os.makedirs(output_dir, exist_ok=True)

# lista med dataframes och titlar
ämnen = ["Engelska", "Matematik", "Svenska", "Svenska som andraspråk"]
dataframes = [df_engelska, df_matematik, df_svenska, df_svenska_som_andraspråk]

# skapa figuren och subplottar
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle("Totala Poäng per Huvudman för olika Ämnen", fontsize=16)

# loopa genom ämnen och plotta
for ax, df, ämne in zip(axes.flatten(), dataframes, ämnen):
    # Filtrera ut relevanta kolumner
    df_filtered = df[["Huvudman", "Totalt (Poäng)"]]
    ax.bar(df_filtered["Huvudman"], df_filtered["Totalt (Poäng)"])
    ax.set_title(ämne, fontsize=14)
    ax.set_xlabel("Huvudman", fontsize=12)
    ax.set_ylabel("Totalt (Poäng)", fontsize=12)
    ax.set_xticklabels(df_filtered["Huvudman"], rotation=45, ha="right")

# anpassa layout
plt.tight_layout(rect=[0, 0, 1, 0.96])

# spara figuren
output_path = os.path.join(output_dir, "totala_poäng_per_huvudman.png")
plt.savefig(output_path, dpi=300)

# visa figuren
plt.show()
```
